```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://cdn.marschannels.com/files/feed/omjwr7pvui.xml'
```

```python
reqs = requests.get(url)
```

```python
soup = BeautifulSoup(reqs.text, 'html.parser')
```

```python
duration = []

for link in soup.find_all('media:content'):
  duration.append(link.get('duration'))  
```

```python
print(duration)
```

```python
# create dataframe

duration = pd.DataFrame({"duration":duration})
```
